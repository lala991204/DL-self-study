```python
# 텐서플로 데이터셋
import tensorflow_datasets as tfds

dataset_name = 'cifar100'
ds = tfds.load(dataset_name, split='train')
ds
```

```python
# 5개의 데이터 출력
for data in ds.take(5):
    print(data)     # 개별 데이터는 딕셔너리 형태로 이루어짐.
```

```python
# 5개의 데이터 출력
for data in ds.take(5):
    # image
    image = data['image']
    #label
    label = data['label']
    print(image.shape)
    print(label)       # 0~99 값들로 이루어짐.
```

다음과 같이 투플 형식으로 dataset 로드 시에 키 값을 확인하지 않아도 되므로 많이 사용되는 옵션이다.

```python
# as_supervised = True는 tuple 형식으로 x, y 데이터 로드
ds = tfds.load(dataset_name, split='train', as_supervised=True)

for image, label in ds.take(5):
    print(image.shape, label)        
```

```python
# with_info = True로 데이터셋의 정보 로드
ds, info = tfds.load(dataset_name, split='train', with_info=True)

# info 확인
info
```
