```python
import tensorflow as tf
import numpy as np

x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]         # 3dim의 데이터를 input하면 (여러개 데이터를 한꺼번에 행렬로 표현)

y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]                  # 1dim의 데이터를 output하는 형태임

tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3))
tf.model.add(tf.keras.layers.Activation('linear'))

tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-5))
tf.model.summary()
history = tf.model.fit(x_data, y_data, epochs=100)

y_predict = tf.model.predict(np.array([[72., 93., 90.]]))       # 3dim 데이터 한 개를 input하면 1dim의 데이터 한 개가 나옴을 알 수 있음.
print(y_predict)
```
