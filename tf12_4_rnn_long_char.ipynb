```python
import tensorflow as tf
import numpy as np

sentence = ("if you want to build a ship, don't drum up people together to "
            "collect wood and don't assign them tasks and work, but rather "
            "teach them to long for the endless immensity of the sea.")

char_set = list(set(sentence))           # index to char
char_dic = {w: i for i, w in enumerate(char_set)}    # char to index

data_dim = len(char_set)        # input dim
hidden_size = len(char_set)     # output dim
num_classes = len(char_set)     # final output dim
sequence_length = 10                                # "Any arbitrary number"
learning_rate = 0.1

dataX = []
dataY = []
# sequence_length 만큼의 window를 움직이면서 수행
for i in range(0, len(sentence) - sequence_length):   
    x_str = sentence[i:i + sequence_length]
    y_str = sentence[i + 1: i + sequence_length + 1]
    print(i, x_str, '->', y_str)

    x = [char_dic[c] for c in x_str]  # x str to index
    y = [char_dic[c] for c in y_str]  # y str to index

    dataX.append(x)
    dataY.append(y)

batch_size = len(dataX)  # 훈련시킬 문자열의 개수

# One-hot encoding
X_one_hot = tf.one_hot(dataX, num_classes)
Y_one_hot = tf.one_hot(dataY, num_classes)

print(X_one_hot.shape)  # check out the shape (170, 10, 25)=>170개의 문자열, 10개의 sequence, 25개의 char
print(Y_one_hot.shape)  # check out the shape


tf.model = tf.keras.Sequential();
tf.model.add(tf.keras.layers.
             LSTM(units=num_classes, input_shape=(sequence_length, X_one_hot.shape[2]), return_sequences=True))
tf.model.add(tf.keras.layers.LSTM(units=num_classes, return_sequences=True))
tf.model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(units=num_classes, activation='softmax')))
tf.model.summary()
tf.model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(lr=learning_rate),
                 metrics=['accuracy'])
tf.model.fit(X_one_hot, Y_one_hot, epochs=100)


results = tf.model.predict(X_one_hot)
print(results)      # 몇 덩어리(문자열 단위)로 존재

for j, result in enumerate(results):
    index = np.argmax(result, axis=1)          # 해당하는 인덱스 추출
    print(index)
    if j == 0:  # print all for the first result to make a sentence
        print(''.join([char_set[t] for t in index]), end='')    # index에 해당하는 char 추출
    else:
        print(char_set[index[-1]], end='')  # 마지막 index부분만 추출한 뒤, 한 칸 띄고
```
